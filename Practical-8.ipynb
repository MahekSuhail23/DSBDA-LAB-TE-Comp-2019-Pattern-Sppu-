```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.jointplot(x='fare',y='age',data=df,hue='who')
mp.show()
```

```python
sns.histplot(df["fare"])
mp.show()
```

0-->false
1-->true

```python
sns.countplot(x ='pclass', data = df,hue='survived')
mp.show()

#passenger of class 1 survived more
```

```python
sns.countplot(x ='sex', data = df,hue='survived')
mp.show()

#women survived are more
```

```python
sns.countplot(x ='sex', data = df,hue='alive', palette = "Set2")
mp.show()

#men rate of alive is low
```
