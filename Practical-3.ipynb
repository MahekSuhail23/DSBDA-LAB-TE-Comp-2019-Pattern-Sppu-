```python
df['Salary']=df['Salary'].astype(float)
df
```

```python
df['Salary'].mean() #for specific col
```

```python
df.mean(axis=1) #row-wise mean
```

```python
print(df.loc[:,'Salary'].mean(skipna=False))
print(df.loc[:,'Salary'].median(skipna=False))
print(df.loc[:,'Salary'].mode())

#loc->access all rows of only salary col to print mean,median,mode
#usually mode is used for qualitative (nominal/ordinal) values
```

```python
print(df.loc[:,'Salary'].min(skipna=False))
print(df.loc[:,'Salary'].max(skipna=False))
```

```python
print('minimum',df['Age'].min())
print('maximum',df['Age'].max())
```

```python
df['Age'].std()
```

```python
print(df.loc[:,'Salary'].std(skipna=False))
print(df.loc[:,'Age'].std(skipna=False))
print(df.loc[:,'Experience_Years'].std(skipna=False))
```
