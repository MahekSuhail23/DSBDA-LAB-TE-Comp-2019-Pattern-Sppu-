Handling outliers: You can handle outliers using statistical methods such as the z-score or interquartile range (IQR) method. The z-score method identifies data points that are more than a specified number of standard deviations from the mean, and the IQR method identifies data points that are outside the range of the middle 50% of the data.

```python
import matplotlib.pyplot as plt
col=['Maths Score','Reading Score','Writing Score']
df.boxplot(col)
```

also called a box-and-whisker plot or a Tukey box plot. 
The center line in the box shows the median for the data
The bottom and top of the box show the 25th and 75th quantiles, or percentiles.
These two quantiles are also called quartiles because each cuts off a quarter (25%) of the data. 
The length of the box is the difference between these two percentiles and is called the interquartile range (IQR).
values that fall above or below the end of the whiskers, they are plotted as dots.
These points are often called outliers

```python
# IQR Tech
sort_maths = nm.sort(df['Maths Score'])
sort_maths
```

```python
Q1 = nm.percentile(df['Maths Score'], 25, interpolation = 'midpoint') 
Q2 = nm.percentile(df['Maths Score'], 50, interpolation = 'midpoint') 
Q3 = nm.percentile(df['Maths Score'], 75, interpolation = 'midpoint') 
 
print('Q1 25 percentile of the given data is, ', Q1)
print('Q1 50 percentile of the given data is, ', Q2)
print('Q1 75 percentile of the given data is, ', Q3)
 
IQR = Q3 - Q1 
print('Interquartile range is', IQR)

low_lim = Q1 - 1.5 * IQR
up_lim = Q3 + 1.5 * IQR
print('low_limit is', low_lim)
print('up_limit is', up_lim)



outlier =[]
for x in df['Maths Score']:
    if ((x> up_lim) or (x<low_lim)):
         outlier.append(x)
print(' outlier in the dataset is', outlier)
```
